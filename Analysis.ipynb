```python
import jsonlines
import numpy as np
```

```python
preds = []
with jsonlines.open('/media/aman/8a3ffbda-8a36-45f9-b426-d146a65d9ece/data1/research/bomm/run_dir_symbolsent_bag3_gumbel_tau10.0_soft_pretrained_frozenwv_5hid_1layer.pred') as pf:
    for line in pf:
        preds.append(line)
```

```python
len(preds)
```

```python
from collections import Counter
print(Counter([x['labels'] for x in preds]))
```

```python
preds[0]
```

```python
group_by_max_loc = {}
for pred in preds:
    arg_max = np.argmax(pred['weights'])
    try:
        group_by_max_loc[arg_max].append(pred['metadata']['text'])
    except KeyError:
        group_by_max_loc[arg_max] = [pred['metadata']['text']]
```

```python
for k,v in group_by_max_loc.items():
    print(k)
    symbolset_1 = sum([1 for x in v if set(x).intersection(set(['0','1','2','3','4','5']))])
    symbolset_2 = sum([1 for x in v if set(x).intersection(set(['a','b','c','d','e','f']))])
    symbolset_3 = sum([1 for x in v if set(x).intersection(set(['x','y']))])
    print(symbolset_1, symbolset_2, symbolset_3)
```
